- PassengerId: ID unik untuk setiap penumpang

- SibSp: Jumlah saudara/kakak/adik dan pasangan di kapal
- Parch: Jumlah orang tua dan anak di kapal
- Ticket, Fare: Informasi tiket dan harga
- Cabin: Nomor kabin (banyak yang kosong)
- Embarked: Pelabuhan tempat naik kapal (C = Cherbourg, Q = Queenstown, S = Southampton)

# **5. Data Preprocessing**

```python
for name, df in [("Train", df_train), ("Test", df_test)]:
    print(f"\nMissing values in {name} set:")
    print(df.isnull().sum())
```

```python
columns_to_drop = ["Name", "Cabin", "PassengerId"]
df_train.drop(columns=columns_to_drop, inplace=True)
df_test.drop(columns=columns_to_drop, inplace=True)
```

```python
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].median())
print(df_train.isnull().sum())
```

```python
mode_embarked = df_train['Embarked'].mode()[0]
df_train['Embarked'] = df_train['Embarked'].fillna(mode_embarked)
print(df_train.isnull().sum())
```

```python
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].median())
print(df_test.isnull().sum())
```

```python
median_fare = df_test['Fare'].median()
df_test['Fare'] = df_test['Fare'].fillna(median_fare)
print(df_test.isnull().sum())
```

```python
def preprocess(df):
    df = df.copy()

    def ticket_number(x):
        return x.split(" ")[-1]

    def ticket_item(x):
        items = x.split(" ")
        if len(items) == 1:
            return "NONE"
        return "_".join(items[0:-1])
    df["Ticket_number"] = df["Ticket"].apply(ticket_number)
    df["Ticket_item"] = df["Ticket"].apply(ticket_item)
    return df

preprocessed_train_df = preprocess(df_train)
preprocessed_serving_df = preprocess(df_test)


preprocessed_train_df.to_csv('../preprocessing/titanic_preprocessing/Titanic_Preprocessed_Train.csv', index=False)
preprocessed_serving_df.to_csv('../preprocessing/titanic_preprocessing/Titanic_Preprocessed_Test.csv', index=False)

preprocessed_train_df.head(5)
```
